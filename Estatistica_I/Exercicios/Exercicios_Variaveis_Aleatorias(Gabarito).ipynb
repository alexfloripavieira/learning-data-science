**Questão 1**

Dentre os 7 atletas, designados por A, B, C, D, E, F e G, com maiores notas de uma prova de skate dois serão sorteados para receber um auxílio financeiro do patrocinador da competição. As notas destes atletas são: 8,7; 8,6; 9,4; 9,6; 8,8; 9,5; 9,2.
    
(A) Defina um espaço amostral para esse experimento e diga a sua cardinalidade.

(B) Seja a variável X = "Nota média dos atletas selecionados". Liste os possíveis valores de $X \geq 9$.

(C) Qual a probabilidade de ser escolhida uma dupla com nota média maior ou igual a 9?

(A) Como a ordem não importa, temos: 
$$\Omega = \{(A,B),(A,C),(A,D),(A,E),(A,F),(A,G),
               (B,C),(B,D),(B,E),(B,F),(B,G),
               (C,D),(C,E),(C,F),(C,G),
               (D,E),(D,F),(D,G),
               (E,F),(E,G),
               (F,G)\}$$
A cardinalidade de $\Omega$ é igual a

$$ \#\Omega = \frac{n!}{p!(n - p)!} = \frac{7!}{2!(7 - 2)!} = \frac{5040}{240} = 21$$

(B) Evento A - $X \geq 9$

$$A = \{(A,C), (A,D), (A,F), (B,C), (B,D), (B,F), (C,D), (C,E), (C,F), (C,G), (D,E), (D,F), (D,G), (E,F), (E,G), (F,G)\}$$

(C)

$$P(X \geq 9) =  \frac{ \#A }{ \#\Omega } = \frac{16}{21} = 0,76 $$

**Questão 2**

O João é um bom lançador de dardos e acerta 35% dos lançamentos na mosca do alvo. Se ele faz 14 lançamentos, qual a probabilidade de ele acertar na mosca no máximo 2 vezes?

Podemos pensar no experimento seguindo uma distribuição Binomial com a probabilidade de sucesso de cada lançamento igual a 0,35. Queremos descobrir $P(X \leq 2)$ onde $X$ é o número de acertos em 14 lançamentos. Assim, temos:

```python
from scipy.stats import binom

prob = binom.pmf(0,14,0.35) + binom.pmf(1,14,0.35) + binom.pmf(2,14,0.35)

print(f'P(X \leq 2) = {prob:.2f}')
```

**Questão 3**

O número médio de clientes que entram em um supermercado por hora é igual a 15. Obtenha a probabilidade de, na próxima hora, entrarem exatamente 21 clientes.

```python
from scipy.stats import poisson
lamb = 15
k = 21
prob = poisson.pmf(k=k,mu=lamb)

print(f'A probabilidade de, na próxima hora, entrarem exatamente 21 clientes é igual a {prob:.2f}')
```

**Questão 4**

A concentração de um poluente em água liberada por uma fábrica tem distribuição N(8; 4). Qual a chance, de que num dado dia, a concentração do poluente exceda o limite regulatório de 10 ppm?

A solução do problema consiste em determinar a proporção da distribuição que está acima de 10 ppm, isto é, $P(X>10)$. Usando o z-score, temos:

$$P(X>10) = P(Z>\frac{10-8}{2}) = P(Z>1) = 1 - P(Z \leq 1)$$

```python
from scipy.stats import norm
import numpy as np

prob = 1 - norm.cdf(1)

print(f'A chance, de que num dado dia, a concentração do poluente exceda o limite regulatório de 10 ppm é igual a {prob:.2f}')
```

Poderíamos executar o cálculo sem transformar para Z:

```python
prob = 1 - norm.cdf(10,loc=8,scale=2)

print(f'A chance, de que num dado dia, a concentração do poluente exceda o limite regulatório de 10 ppm é igual a {prob:.2f}')
```
